```
def kernel2fft(nu, n1, n2, separable=None):
    import itertools
    if separable is None:      
        s1 = (nu.shape[0] - 1) // 2
        s2 = (nu.shape[1] - 1) // 2

        assert(n1>=s1 and n2>=s2)

        tmp = np.zeros((n1, n2))
        tmp[:s1+1, :s2+1] = nu[s1:2*s1+1, s2:2*s2+1]
        #yellow
        tmp[n1-s2:, :s2+1] = nu[:s1, s2:2*s2+1]
        # green
        tmp[:s1+1, n2-s2:] = nu[s1:2*s1+1, :s2]
        # red
        tmp[n1-s1:, n2-s2:] = nu[:s1, :s2]
    else:
        def helper_sep(nu1,nu2,idx1,idx2):   
            temp = np.zeros((idx1[1]-idx1[0],idx2[1]-idx2[0]))
            
            if separable=='product':
                temp = np.matmul(nu1[slice(*idx1)],nu2[:,slice(*idx2)])
            elif separable=='sum':
                indeces = itertools.product(range(idx1[0],idx1[1]), \
                                            range(idx2[0],idx2[1]))
                for (id1,id2) in indeces:
                    i,j = id1-idx1[0],id2-idx2[0]
                    if (id2==0):
                        temp[i,j] += nu1[id1]
                    if (id1==0):
                        temp[i,j] += nu2[:,id2]
                    
            return temp
        
        nu1, nu2 = nu
        s1 = (nu1.size - 1) // 2
        s2 = (nu2.size - 1) // 2
        assert(n1>=s1 and n2>=s2)

        tmp = np.zeros((n1, n2))
        tmp[:s1+1, :s2+1] = helper_sep(nu1,nu2,(s1,2*s1+1), (s2,2*s2+1))
        #yellow
        tmp[n1-s2:, :s2+1] = helper_sep(nu1,nu2,(0,s1), (s2,2*s2+1))
        # green
        tmp[:s1+1, n2-s2:] = helper_sep(nu1,nu2,(s1,2*s1+1), (0,s2))
        # red
        tmp[n1-s1:, n2-s2:] = helper_sep(nu1,nu2,(0,s1), (0,s2))
    
    lbd = npf.fft2(tmp,axes=(0,1))

    return lbd
```

Code is provided here for convenience

```
def convolvefft(x, lbd):
    fft_x = npf.fft2(x,axes=(0,1))
    
    # if lambda kernel is 2D, extend to 3D
    if lbd.ndim==2 and x.ndim==3:
        # https://stackoverflow.com/questions/32171917/copy-2d-array-into-3rd-dimension-n-times-python
        kernel = np.broadcast_to(lbd[...,None],lbd.shape+(3,))
    elif lbd.ndim==2 and x.ndim==2:
        # x is greyscale
        kernel = lbd
    else:
        # Assume that lbd.ndim is 2D
        raise NotImplementedError
               
    result = fft_x * kernel
    result = npf.ifft2(result,axes=(0,1))
    
    return np.real(result)
```

## Q10

````python
x = plt.imread('assets/lady.png')        

def test_convolvefft(x,tau,separable=None,type_kernel=0):
    # type_kernel is the index for the following list
    kernels = ['gaussian','exponential','box']
    kernel = kernels[type_kernel]
        
    # preprocessing steps    
    if separable is None:
        nu = im.kernel(kernel,tau)
    else:
        nu1 = im.kernel(kernel+'1', tau)
        nu2 = im.kernel(kernel+'2', tau)
        nu = nu1,nu2
    
    # test time
    lbd = im.kernel2fft(nu,x.shape[0],x.shape[1],separable=separable)
    time_fft = time.time()
    result_fft = im.convolvefft(x,lbd)
    time_fft = time.time() - time_fft
    
    time_conv = time.time()
    result_conv = im.convolve(x,nu,separable=separable)
    time_conv = time.time() - time_conv
    
    # if tau is zero, they will be within atol=1e-6 of each other. This is not much. 
    return np.allclose(result_fft,result_conv,atol=1e-6),time_fft,time_conv
````

````python
# test for that convolve. Runs through every type of kernel and every type of seperation 

def test_convolve(tau):
    test = []
    for type_kernel in range(3):
        for separable in [None,'sum','product']:
            is_close,_,_ = test_convolvefft(x,tau=tau,separable=separable,type_kernel=type_kernel)
            test.append(is_close)
            
    test_result = np.array(test).all()
    
    return test_result
````

## The error between the two is zero. See test_convolvefft above which I use for Q11 as well.

````python
# test for that convolve. Runs through every type of kernel and every type of seperation 

for tau in range(1,5):
    print(test_convolve(tau=tau))
````

### The complexity of the spectral convolution is O(nlogn) + O(n) + O(nlog n) which is O(n) (see lecture slides, this is FFT, multiply, and IFFT.
### The complexity of spatial when separable should be k*O(n) (k=3 or 2 depending on if we coun't multiply) when separable and when not separable should be O(n^2) (see k,l loops). My graph doesn't show this but separable saves 1 loop.
### Spectral is favorable for tau above 1, spatial otherwise.

````python
def plot_results(x):   
    ys_sep = []
    ys = []
    
    # taus
    xs = np.arange(0,5)
    
    
    for tau in xs:
        time_fft , time_conv , sep_time_conv , sep_time_fft = 0 , 0 , 0 , 0
        for i in range(100):           
            for separable in [None,'product','sum']:
                # type_kernel=2 for box
                isclose,t1,t2 = test_convolvefft(x,tau=tau,separable=separable,type_kernel=2)
                if not isclose:
                    raise Exception

                if separable is None:
                    time_fft += t1
                    time_conv += t2
                else:
                    sep_time_fft += t1
                    sep_time_conv += t2
        # average time over 100 runs (product and sum are averaged over each other)
        time_fft /= 100
        time_conv /= 100
        sep_time_fft /= 200
        sep_time_conv/= 200
        
        # append time in seconds 
        ys.append((time_fft,time_conv))
        ys_sep.append((sep_time_fft,sep_time_conv))
    
    return xs,ys,ys_sep
````

````python
x = plt.imread('assets/lady.png')        

xs,ys,ys_sep = plot_results(x)
````

````python
def to_y(results,fft_or_conv,is_separable=None):
    # used to parse results for plotting
    # averages over product and sum for separable
    if 'fft' in fft_or_conv:
        plot_ys = np.array([y[0] for y in results]) * 1000
    else:
        plot_ys = np.array([y[1] for y in results]) * 1000
        
    return plot_ys
````

## Q14,Q15

#### From the matrix form of H and it's structure, given in lecture slides, it is easy to verify that since H* = conjugate(H.T), mu is the time reversed conjugate of nu. since nu is real, mu is just the time reversed nu.


#### citation: pg. 209 and 210 https://books.google.ch/books?id=KV03BAAAQBAJ&pg=PA210&lpg=PA210&dq=adjoint+convolution+-x,-y&source=bl&ots=O0SKLLeIky&sig=ACfU3U3OvtorSEO0SkMMWsEes23iR4V0Tw&hl=en&sa=X&ved=2ahUKEwjO_ZSlsJriAhXQDewKHX-KC04Q6AEwAnoECAYQAQ#v=onepage&q=adjoint%20convolution%20-x%2C-y&f=false

In order to compute this, I used the following reasoning


Take the fourier transform of both sides then take the inverse fourier transform.
$$ F^{-1} F mu = F^{-1} F nu$$

By the conjugate property of fourier transforms, we can find mu in the following way:

$$mu = F^{-1}(conjugate(F(nu)))$$
and the fourier transform of mu as

$$F mu = conjugate(F(nu))$$

````python
nu = im.kernel(name='motion')
x = plt.imread('assets/house.png')
y = plt.imread('assets/map.png')
````

````python
def dot_product(A,B):
    return np.sum(A*B)
````

### Verification for Q14

````python
mu = npf.ifft2(np.conjugate(npf.fft2(nu)))
mu = np.real(mu)
lhs,rhs = dot_product(im.convolve(x,nu),y),dot_product(x,im.convolve(y,mu))
````

````python
np.isclose(lhs,rhs,rtol=.001),lhs,rhs
````

````python
nu_fft = im.kernel2fft(nu,x.shape[0],x.shape[1])
mu_fft = np.conjugate(nu_fft)
lhs,rhs = dot_product(im.convolvefft(x,nu_fft),y),dot_product(x,im.convolvefft(y,mu_fft))
````
